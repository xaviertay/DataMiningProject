```python
import glob
import numpy as np
import pandas as pd
```

```python
path = r'D:\Desktop\Data Mining\structured' # use your path
all_files = glob.glob(path + "/*.csv")
```

```python
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
export_csv = frame.to_csv (r'D:\Desktop\Structured_Data.csv', index = None, header=True)
```
